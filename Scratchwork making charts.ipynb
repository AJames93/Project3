```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
AllStates_df = pd.read_csv("Data.csv", index_col=0)
CA_df=pd.read_csv("CA_data.csv", index_col=0)
NY_df=pd.read_csv("NY_data.csv", index_col=0)
NJ_df=pd.read_csv("NJ_data.csv", index_col=0)
TX_df=pd.read_csv("TX_data.csv", index_col=0)
WA_df=pd.read_csv("WA_data.csv", index_col=0)
```

```python
Pie_chart=WA_df["Type"].value_counts()[:10]

Pie_chart.plot(kind="pie", y='States', title="Top 10 Restaurant Types for Washington", startangle=280, figsize=(12,6), autopct = "%1.1f%%")
plt. ylabel("")
```

```python
weightedAvgs =pd.DataFrame({
    "States":["Califorina", "New Jersey", "New York", "Texas", "Washington"], 
    "Weighted Average":[4.543280775, 4.541929069, 4.518968301, 4.516566012, 4.527777778]
})
weightedAvgs.head()
```

```python
Bar_chart=weightedAvgs.plot.bar(ylim=(4.51,4.55), rot='vertical',title='Weighted Averages of Top 20 Per State')
Bar_chart.set_xlabel("States")
Bar_chart.set_ylabel("Weighted Averages") 
Bar_chart.set_xticklabels(["CA", "NJ", "NY", "TX", "WA"])
rects = Bar_chart.patches
labels = [4.543281,4.541929,4.518968,4.516566,4.527778]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    Bar_chart.text(
        rect.get_x() + rect.get_width() / 2, height, label, ha="center", va="bottom"
    )
```
